```python
import pandas as pd
import numpy as np
```

```python
data_may_04_19 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_190504.txt")
data_may_11_19 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_190511.txt")
#data_may_18_19 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_190518.txt")
#data_may_25_19 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_190525.txt")
#data_may_05_18 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_180505.txt")
#data_may_12_18 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_180512.txt")
#data_may_19_18 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_180519.txt")
#data_may_26_18 = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_180526.txt")
```

```python
all_data = pd.concat([data_may_04_19, data_may_11_19], ignore_index=True)
#all_data = pd.concat([data_may_04_19, data_may_11_19, data_may_18_19, data_may_25_19, data_may_05_18, data_may_12_18, data_may_19_18, data_may_26_18])
all_data.info()
```

```python
all_data['datetime'] = pd.to_datetime(all_data['DATE'] + ' ' + all_data['TIME'])
```

```python
all_data['turnstile'] = all_data['C/A'] + all_data['UNIT'] + all_data['SCP'] + all_data['STATION']
```

```python
all_data.sort_values(by = ['turnstile', 'datetime'], inplace = True)
```

```python
all_data.tail(50)
```

```python
#all_data['first_datetime'] = all_data.groupby('turnstile').min('datetime')

# I used 'First' as fillna so it would be easier to see how many it would change, will sub in 0 later
all_data['test'] = all_data.groupby('turnstile')['ENTRIES'].diff().fillna('First')
all_data['test'].value_counts()
```
